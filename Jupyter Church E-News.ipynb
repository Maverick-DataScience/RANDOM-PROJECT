# DATA PROCESSING for Central Baptist Church Survey

In this project, I will try to process and analyze the data that has been collected from the survey conducted in Central Baptist Church at 11th December and 18th December 2022. I will try to group the interest by age group and see which news source the particular age group accessed the most and the number of news each age group read. I will also try to find which news messages people think aren't delivered well. For this question in particular there's no need to group it by age group.

While the number of sample is not big enough to represent the whole International Service community, we assume that the sample itself already can present the community.

```python
import textwrap
plt.figure(figsize=(14,5))
plt.xlim(0, 15)
plt.xlabel("Numbers of Vote", size=15, fontweight='bold')
y=[0,1,2,3,4]

text=['Pastor Desk','Weekly Announcement Details (such as Contact Person, Contact Number, Date & Time, Venue, etc.)','Church Ministry Details (Life Groups, Worship, Welcoming, etc.)','Church Meeting Details', 'Weekly Announcements (Events, Messages, etc.)']
sorted_data = data_split_2.groupby(["What news messages from below do you think aren't delivered well?"])["What news messages from below do you think aren't delivered well?"].agg('count').reset_index(name='Count')

plt.barh(y, sorted_data['Count'])
plt.yticks(y, [textwrap.fill(t, 30) for t in text], fontsize = 15, horizontalalignment='right', fontweight='bold')
plt.title("Information That Is Not Clear Enough", size = 30, fontweight = 100, pad=30, loc='center')

```

##### Based from the graph above, lot of people think Church Ministry Details still haven't been delivered well. Several people think that other information from the survey are also isn't conveyed properly.

```python
from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder(dtype = int)
enc
```

```python
data['Age_Group'] = enc.fit_transform(data[['What age group are you in?']])
data['News_Source'] = enc.fit_transform(data[['Where do you usually receive Church News from? (maximum 2)']])
data['No_News_Read'] = enc.fit_transform(data[['In the past 2 months, how many Weekly E-News Bulletins have you read?']])
data.columns.tolist()
```
